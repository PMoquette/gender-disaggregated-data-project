### Source Headline: playing with Textblob

Some of the headlines and sources are in French. I don't yet know how many, but it would be nice to be able to use them. I'll be using the Textblob and langdetect libraries to take a closer look at these headlines.

```python
from textblob import TextBlob
from langdetect import detect
import langid
```

```python
cod_data.source_headline = cod_data.source_headline.fillna('')
```

```python
def english_french(text):
    langid.set_languages(['fr', 'en'])
    try:
        lang, score = langid.classify(text)
        return lang
    except:
        return None

cod_data['language'] = cod_data['source_headline'].apply(english_french)
```
